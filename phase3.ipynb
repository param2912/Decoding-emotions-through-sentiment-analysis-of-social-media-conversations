```python
from textblob import TextBlob

# Sample social media posts
posts = [
    "I love the new design of the app!",
    "This update ruined everything. So frustrating.",
    "Meh, it's okay. Not great, not terrible.",
    "Absolutely fantastic service today!",
    "I'm really disappointed with the customer support."
]

# Analyze sentiment of each post
for post in posts:
    analysis = TextBlob(post)
    polarity = analysis.sentiment.polarity
    if polarity > 0:
        sentiment = "Positive"
    elif polarity < 0:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    print(f"Post: {post}\nSentiment: {sentiment}\n")
```
